```python
with open(r"../part1.md", "r") as f:
    part1_task = f.read().strip()
print(part1_task)
with open(r"../part2.md", "r") as f:
    part2_task = f.read().strip()
print(part2_task)
with open(r"../day1_part1_2024.go", "r") as f:
    part1_solution = f.read().strip()
print(part1_solution)
with open(r"../day1_part2_2024.go", "r") as f:
    part2_solution = f.read().strip()
print(part2_solution)
with open(r"../input.txt", "r") as f:
    input = f.read().strip()
part1_answer = "2430334"
part2_answer = "28786472"
```

```python
new_items = [
    {
        "name": "day1_part1_2024",
        "task": part1_task,
        "input": input,
        "answer": part1_answer,
        "solution": part1_solution,
        "solution_lang": "go",
        "year": 2024
    },
    {
        "name": "day1_part2_2024",
        "task": part2_task,
        "input": input,
        "answer": part2_answer,
        "solution": part2_solution,
        "solution_lang": "go",
        "year": 2024
    },
]
```

```python
import json

# Load the file
file_path = '../../../git/advent-of-code/train.json'
with open(file_path, 'r') as file:
    data = json.load(file)

# Append the new items
data.extend(new_items)

# Save the updated data back to the file without extra formatting
with open(file_path, 'w') as file:
    json.dump(data, file)
```
